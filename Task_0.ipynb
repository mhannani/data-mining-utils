# Distance between true-color, gray-scale and binary images

```python
# Using tf-gpu local conda environment
import numpy as np
from distance.distances import Distances
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
plt.imshow(X_train[19])
```

```python
plt.imshow(X_train[0]);
```

## 1. Measuring distances between true-color images

* #### Euclidean distance

$$Euclidean(x, y) = \sqrt{\sum_{i=1}^n (x_i-y_i)^2}   $$

```python
Distances.euclidean(X_train[0], X_train[19])
```

* #### Manhattan distance

The Manhattan distance (sometimes also called Taxicab distance) metric is related to the Euclidean distance, but instead of calculating the shortest diagonal path ("beeline") between two points, it calculates the distance based on gridlines. The Manhattan distance was named after the block-like layout of the streets in Manhattan.

$$Manhattan(x, y) = \left(\sum_{i=1}^n |x_i-y_i|^p\right)^{1/p}$$

```python
Distances.manhattan(X_train[0], X_train[19])
```

* #### Minkowski distance

The Minkowski distance is a generalized form of the Euclidean distance (if p=2) and the Manhattan distance (if p=1)

$$Minkowski(x, y) = \left(\sum_{i=1}^n |x_i-y_i|^p\right)^{1/p}$$

```python
Distances.minkowski(X_train[0], X_train[19], 4)
```
